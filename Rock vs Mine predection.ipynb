```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('sonar_data.csv', header=None)
```

```python
data.head()
```

```python
data.isnull().count()
```

```python
data.dropna()
```

```python
data.isnull().count()
```

```python
data.shape
```

```python
X = data.drop(60,axis = 1)
Y = data[60]
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.87,stratify = Y, random_state = 12)
```

```python
print(X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
accuracy_train = accuracy_score(X_train_prediction, Y_train)
```

```python
print('accuracy on training data is',accuracy_train)
```

```python
X_test_prediction  = model.predict(X_test)
accuracy_test = accuracy_score(X_test_prediction, Y_test)
```

```python
print('accuracy on test data is ',accuracy_test)
```
